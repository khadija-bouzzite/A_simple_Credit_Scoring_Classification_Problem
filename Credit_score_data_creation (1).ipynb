# Creating random data

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
sectors_exmpl=['Consultant',
  'Employment Office',
  'Software',
  'Entertainment',
  'Delivery',
  'Fundraisers',
  'Cosmetics',
  'Healthcare',
  'Platform',
  'Marketing',
  'Notary Public',
  'Online Business',
  'Publishing Services',
  'Record Business',
  'Retail Sales',
  'Technology Services',
  'Telemarketing',
  'Travel Agency',
  'Video Production',
  'Architect',
  'Building Construction',
  'Building Inspection',
  'Concrete Manufacturing',
  'Contractor',
  'Engineering',
  'Equipment Rental',
  'Plumbing',
  'Remodeling',
  'Repair',
  'Maintenance',
  'Dentistry',
  'Electrolysis',
  'Embalmer',
  'Emergency Medical Services',
  'Hearing Aid Dealers',
  'Home Health Services',
  'Farming',
  'Fishing',
  'Landscape']
sector=[]
for i in range(500):
    sector.append(random.choice(sectors_exmpl))

Credit_score=[0,10]
target=[]
for i in range(500):
    target.append(random.choice(Credit_score))
    
Mouvement=["Pr","Rg","Nat","Int"]
M=[]    
for i in range(500):
    M.append(random.choice(Mouvement))

Siege=["Casa","Rabat","Laayoune","Marrakech","Internet","None"]
S=[]    
for i in range(500):
    S.append(random.choice(Siege))
 
    
```

```python
dict1={ "revenue":random.choices(range(1,1000000), k=500),
       "Nb_cred":random.choices(range(0,1000), k=500),
       "Cred_in_prog":random.choices(range(1,10),k=500),
       "Chiffre_daf":random.choices(range(1,1000000),k=500),
       "secteur":sector,
       "Mouv_geo":M,
       "Internet":random.choices(range(0,1),k=500),
       "Inves_extern":random.choices(range(0,1),k=500),
       "site_web":random.choices(range(0,1),k=500),
       "Covid_impact":random.choices(range(0,1),k=500),
       "sponsor":random.choices(range(0,1),k=500),
       "Nb_Abb":random.choices(range(0,1000000), k=500),
       "Siege":S,
       "Credit_score":target


       }
      
```

```python
Data=pd.DataFrame(dict1)
```

```python
Data["Nb_cred"]=Data["Nb_cred"]/200
```

```python
Data["Nb_cred"]=Data["Nb_cred"].astype(int)
```

```python
Data["Nb_cred"]=Data["Nb_cred"]+Data["Cred_in_prog"]
```

```python
Taux_sat=[0,50,100]
Ts=[]    
for i in range(500):
    Ts.append(random.choice(Taux_sat))
```

```python
Data["Taux_sat"]=Ts
```
